```python
import pandas as pd
data = {
    "Tanggal":["2025-01-02","2025-01-03","2025-01-04","2025-01-05","2025-01-06","2025-01-07","2025-01-08","2025-01-09","2025-01-09","2025-01-10",],
    "Produk":["Mochi choco","Mochi stroberi","Mochi tiramisu","Mochi keju","Donat","Ice Cookies","Cookies coklat","Cookies keju","Cookies stoberi","Cookies marsmellow"],
    "Jumlah terjual":[14,10,20,4,12,20,15,23,10,5],
    "Stok":[1,23,15,20,14,10,11,18,12,10],
    "Harga satuan":[15000, 16500, 14000, 18000,10000,20000,25000,10000,17000,20000],
}

dt = pd.DataFrame(data)
dt
```

```python
dt.info()
```

```python
dt["Tanggal"] = pd.to_datetime(dt["Tanggal"])

dt["Total Penjualan"] = dt ["Jumlah terjual"] * dt["Harga satuan"]
dt["Keuntungan"] = (dt["Total Penjualan"]-dt["Jumlah terjual"]*10000)
dt
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

x = dt[["Jumlah terjual","Stok"]]
y = (dt["Stok"] <5)

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2, random_state=42)

model = DecisionTreeClassifier()
model.fit (x_train,y_train)
```

```python
y_pred = model.predict(x_test)
print("Akurasi model:",accuracy_score(y_test,y_pred))
```

```python
produk_baru = [[8, 6]]
predict = model.predict(produk_baru)

if predict[0] == 1:
    print("Perlu di restock!")
else:
    print("Stok masih ada")
```
